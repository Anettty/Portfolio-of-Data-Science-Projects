```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = pd.read_csv("titanic.csv")
titanic
```

```python
sns.jointplot(x="fare", y="age", data= titanic, color="dodgerblue", alpha=0.8)
```

```python
titanic["fare"].hist(color="orange")
```

```python
sns.distplot(titanic['fare'], kde=False, bins=40, color="red")
```

```python
sns.boxplot(x="class", y="age", data=titanic, palette="pastel")
```

```python
sns.swarmplot(x='class', y="age", data=titanic, palette="Set1")
```

```python
sns.countplot(x="sex", data=titanic)
```

```python
hm=titanic.corr()
sns.heatmap(hm, cmap="coolwarm")
plt.title("titanic.corr()")
```

```python
g = sns.FacetGrid(titanic, col="age",  row="sex")
g = g.map(plt.hist, "age")
g
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g= sns.FacetGrid(data=titanic, col= "sex")
g.map(sns.distplot, "age", kde=False)

```
