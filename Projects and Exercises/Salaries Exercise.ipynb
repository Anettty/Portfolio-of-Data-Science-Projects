** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
df_list = sal["JobTitle"].tolist()
df_list

```

```python
new=[]
for n in df_list:
    if n.lower not in new and "chief" in n.lower():
        new.append(n)
len(new)
```

```python
new1= []
for f in new:
    if "chief" in f.lower():
        new1.append(f)
len(new1)
```

```python
def thereisone(t):
    if "chief" in t.lower():
        return True
    else:
        return False
sum(sal['JobTitle'].apply(lambda x: thereisone(x)))
```
